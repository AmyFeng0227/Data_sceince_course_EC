# About databases

The trainng of face expression used a combination of two databases: 
1. The Face expression recognition dataset from Kaggle (https://www.kaggle.com/datasets/jonathanoheix/face-expression-recognition-dataset): gray pictures
2. Real-world Affective Faces Database (RAF-DB) (http://www.whdeng.cn/RAF/model1.html): mix of gray and color pictures

The training of age, gender and race used database UTKFace (https://susanqq.github.io/UTKFace/): each image with annotations of age, gender, and ethnicity
